```python
data = flight_data
# remove categorical columns
data = data.drop(columns=['month', 'year'], axis=1)

# train-test split
def train_test_split(data, split=0.8):
    """Splits dataset into train-test splits.

    Args:
        data (pd.DataFrame): The dataset.
        split (float, optional): Proportion of the 
            dataset used for training. Defaults to 0.8.

    Returns:
        (np.ndarray): Training set.
        (np.ndarray): Testing set.
    """ 
    data = np.array(data)
    train_size = int(split * data.shape[0])
    data_train = data[:train_size]
    data_test = data[train_size:]
    return data_train, data_test

train_data, test_data = train_test_split(data)

# scale training set for training 
scaler = MinMaxScaler(feature_range=(-1, 1))
train_data = scaler.fit_transform(train_data .reshape(-1, 1))
test_data = scaler.fit_transform(test_data .reshape(-1, 1))


# conve
print(train_data.shape, test_data.shape)

# def sliding_windows(data, sequence_length):
#     f
```
